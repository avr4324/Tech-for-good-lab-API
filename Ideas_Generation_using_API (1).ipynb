```python
import openai

# Defining an API_call function
def API_call(model, OPENAI_KEY, prompt, system_prompt=''):

    openai.api_key = OPENAI_KEY

    response = openai.ChatCompletion.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ]
    )
    return response['choices'][0]['message']['content']
```

```python
import openai
import numpy as np

api_key = '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'

def get_openai_embedding(text, api_key):
    response = openai.Embedding.create(
        input=text,
        model="text-embedding-ada-002",  # Use the appropriate model name
        api_key=api_key
    )
    return response['data'][0]['embedding']

# Fetch embeddings for all texts
embeddings = [get_openai_embedding(text, api_key) for text in combined_texts]
embeddings_array = np.array(embeddings)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from scipy.spatial.distance import cdist
pca = PCA(n_components=2)
pca_result = pca.fit_transform(embeddings_array)

# Create a DataFrame for plotting
df = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
df['label'] = [f"Note {i+1}" for i in range(len(embeddings_array))]

# Calculate mean and standard deviation of embeddings
mean_vector = np.mean(embeddings_array, axis=0)
std_vector = np.std(embeddings_array, axis=0)

# Perform PCA on mean and standard deviation vectors
mean_pca = pca.transform([mean_vector])[0]
std_pca = pca.transform([std_vector])[0]

# Calculate the radius for the circle (use norm of std_pca to get a single radius value)
radius = np.linalg.norm(std_pca) * 2

# Calculate distances between points in PCA space
distances = cdist(pca_result, pca_result)

# Find nearest prior point for each point
nearest_prior_indices = []
nearest_prior_distances = []
for i in range(1, len(distances)):
    prior_distances = distances[i, :i]
    nearest_prior_idx = np.argmin(prior_distances)
    nearest_prior_distances.append(np.min(prior_distances))
    nearest_prior_indices.append(nearest_prior_idx)

# Add a zero for the first point which has no prior points
nearest_prior_distances = [0] + nearest_prior_distances

# Add the nearest prior distances to the DataFrame
df['nearest_prior_distance'] = nearest_prior_distances

# Plotting
plt.figure(figsize=(12, 10))
plt.scatter(df['PCA1'], df['PCA2'], alpha=0.7, label='Post-it notes')
```
